## Creamos los gráficos para ver las proporciones categórica en cada categoría segmento de ingreso.

```python
#definimos una función que nos permite darle la categoría como argumento y que nos devuelva los ratios.
df_prueba10 = pd.DataFrame()
def pct (category):
    df_prueba10 = df.groupby([category,"income"]).agg({"income":"count"})
    return pd.DataFrame(df_prueba10.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))).unstack(level =1)
```
