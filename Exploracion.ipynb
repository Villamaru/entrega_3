## Creamos los grÃ¡ficos para ver las proporciones categÃ³rica en cada categorÃ­a segmento de ingreso.

```python
#definimos una funciÃ³n que nos permite darle la categorÃ­a como argumento y que nos devuelva los ratios.
df_prueba10 = pd.DataFrame()
def pct (category):
    df_prueba10 = df.groupby([category,"income"]).agg({"income":"count"})
    return pd.DataFrame(df_prueba10.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))).unstack(level =1)
```
